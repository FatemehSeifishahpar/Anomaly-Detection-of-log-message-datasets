```python
# get positive from pu and label them 1
x_train_out_pu_pos = x_train_df_pu.loc[out_pu.index]
y_train_out_pu_pos = y_train_df_pu.loc[out_pu.index]
# x_train_out_pu_pos['label'] = y_train_out_pu_pos
x_train_out_pu_pos['label'] = 1
x_train_out_pu_pos
```

```python
# get negative from pu and label them 0

y_train_out_pu_neg = y_train_df_pu[y_train_df_pu == 0]
x_train_out_pu_neg = x_train_df_pu.loc[y_train_out_pu_neg.index]
x_train_out_pu_neg['label'] = y_train_out_pu_neg
x_train_out_pu_neg
```

```python
train = pd.concat([x_train_out_pu_neg, x_train_out_pu_pos])
train = train.sample(frac=1.0)
train
```

```python
X_train = train.iloc[:, :x_train_df_pu.shape[1]]
y_train = train['label']
X_train
```

```python
model = Sequential()
model.add(Embedding(2000000, 10, input_length=X_train.shape[1]))
model.add(LSTM(128, dropout=0.2, recurrent_dropout=0.2))
model.add(Dense(1, activation='sigmoid'))

# try using different optimizers and different optimizer configs
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

print('Train...')
model.summary()
model.fit(X_train.to_numpy(), y_train.to_numpy(),
          batch_size=32,
          epochs=3,
          validation_split=0.1)
# score, acc = model.evaluate(x_test_df.to_numpy(), y_test_df.to_numpy(),
#                             batch_size=128)
# print('Test score:', score)
# print('Test accuracy:', acc)
```

```python
y_pred = model.predict(X_test.to_numpy())
```

```python
y_pred = (y_pred > 0.5)
```

```python
matrix = metrics.confusion_matrix(y_test["truth_"].values, y_pred)
matrix
```
